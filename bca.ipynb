```python
from bs4 import BeautifulSoup
import requests
import yfinance as yf
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
bca = yf.Ticker("BBCA")
```

```python
bca_data = bca.history(period="max")
```

```python
bca_data.head()
```

```python
bca_data.reset_index(inplace=True)
bca_data
```

```python
bca_data.tail()
```

```python
bca_data.sort_values(by='Date', ascending=False, inplace=True)
bca_data
```

```python
bca_data.reset_index(drop=True, inplace=True)
bca_data
```

```python
bca_data['Date'] = pd.to_datetime(bca_data['Date'], utc=True).dt.date
bca_data['Date']
```

```python
bca_data
```

```python
bca_data.index = bca_data['Date']
bca_data = bca_data.drop(columns='Date')
```

```python
bca_data
```

```python
bca_data['Dividends'].unique()
bca_data['Dividends'].value_counts()
```

```python
bca_data['Stock Splits'].unique()
bca_data['Stock Splits'].value_counts()
```

```python
bca_data['Capital Gains'].unique()
bca_data['Capital Gains'].value_counts()
```

```python
bca_data.drop(columns=['Dividends', 'Stock Splits', 'Capital Gains'], inplace=True)
```
